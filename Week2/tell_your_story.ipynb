### **What to Do This Week**

**Step 1: List Possible Tables**
Create a short list of possible tables your project might include.

* **For each table, include:**
  * A table name (informal is fine)
  * A short description of what it represents
  * A few example fields (column ideas)

* **Examples Include:**
  * **Cities**: location information like name, state, region
  * **Weather Readings**: date, temperature, rainfall
  * **Air Quality**: date, PM2.5 value, location
  
This is a draft — nothing needs to be perfect.

### **MY TABLES**

**Step 2: Sketch the Connections**

* **Using paper, a whiteboard, sticky notes, or a simple digital tool:**
  * Draw boxes for each table idea
  * Draw lines between tables that seem related
  * Add a note describing how they connect (shared city, shared date, etc.)

A rough sketch is completely acceptable.

### **Step 3: Explain Your Thinking**
* **Write a short paragraph (5–7 sentences) explaining:**
  * What your main tables represent
  * How your two datasets connect conceptually
  * Why this structure makes sense for the questions you want to explore

Use plain language. Pretend you are explaining your project to a classmate, not a database expert.

### **MY THINKING**

* **Important Notes:**
  * You are not expected to define data types or keys.   
  * You are not expected to get this “right” on the first try.   
  * You will revise and refine this structure after learning formal database design later in the course.   

### **This Week’s Goal**
* By the end of this week, you should be able to say:
  * “I can explain the main pieces of my project and how they relate, even if I haven’t built anything yet.”

That understanding will make the technical work much easier when you get there.
