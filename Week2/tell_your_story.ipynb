### **Tables**

* Possible Tables:
  * Satellites: core table of all tracked objects:
    * norad_id (PK), object_name, object_type, launch_date, ops_status, owner_id, launch_id
  * orbital_data: physics, geometry, and structural fields for each satellite:
    * norad_id (PK/FK), orbit_class, inclination_degrees, period_minutes, perigee_km, apogee_km, eccentricity
    * mass_kg, dry_mass_kg, rcs_m2, rcs_class, mass_source_tier
  * ownership_operators: country and operator metadata (who owns/operates each payload):
    * owner_id (PK), owner_country, operator_name, users_category, purpose
  * launch_events: launch information (one launch can produce many catalog objects):
    * launch_id (PK), launch_date, launch_site, launch_vehicle, launch_year
  * ucs_details: UCS descriptive details for each satellite record:
    * norad_id (PK/FK), satellite_name, lifetime_years, launch_mass_kg, orbit_type, active_flag
  * risk_assessment: derived analytics (velocity, kinetic energy, zombie flag, risk band):
    * norad_id (PK/FK), mean_orbital_velocity_kms, kinetic_energy_index, zombie_flag, risk_band

### **Step 2 Notes: How the Tables Connect**

![Schema Design](./schema_design.png)

* **satellites ↔ orbital_data**
  * Connected by `norad_id` (shared satellite identifier).
  * Use this join to connect identity/status fields with orbit and physical attributes.

* **satellites ↔ ucs_details**
  * Connected by `norad_id`.
  * Adds UCS-maintained fields (active flag, lifetime, launch mass) to each satellite.

* **satellites ↔ risk_assessment**
  * Connected by `norad_id`.
  * Links each satellite to derived risk measures like kinetic energy index and risk band.

* **satellites ↔ ownership_operators**
  * Connected by `owner_id`.
  * Many satellites can map to one owner/operator profile.

* **satellites ↔ launch_events**
  * Connected by `launch_id`.
  * Many satellites can map to one launch event (same vehicle/site/date).

### **Step 3: Explain Your Thinking**

I built this project around the `satellites` table because `norad_id` gives each satellite one unique ID and keeps the core details in one place. Then I used `orbital_data`, `ucs_details`, and `risk_assessment` as support tables that add more detail to that same satellite record. I kept ownership and launch in separate lookup tables so I do not have to repeat owner or launch info in every satellite row. Instead, I can store `owner_id` and `launch_id` in `satellites` and join when I need those details. This setup makes it easier to answer different questions because I can pull only the columns I need for each analysis. Overall, the structure reduces redundancy and keeps the data organized while still supporting questions about status, mission purpose, and risk patterns.
