What to Do This Week
====================

### Step 1: Create a Draft Table Layout

Create a document or diagram that includes:

*   All draft tables
    
*   Columns for each table
    
*   Notes about how tables connect
    

This can be:

*   A hand-drawn sketch (photo or scan is fine)
    
*   A simple digital diagram
    
*   A structured text document with headings
    

Clarity matters more than format.

Step 2: Write a Short Explanation
---------------------------------

Write a short explanation (1–2 paragraphs) that answers:

*   What does each table represent?
    
*   How do the tables connect?
    
*   How does this structure support the questions you want to answer?
    

Avoid technical language. Focus on meaning and logic.

Important Reminders
-------------------

*   This is **still a draft**, not a final schema.
    
*   You do not need to define data types or constraints.
    
*   You do not need to be “correct” — you need to be thoughtful.
    
*   You will revise this after learning formal database design.
    

This Week’s Goal
----------------

By the end of this week, you should be able to say:

### _“I have a clear draft of how my data could be organized into tables, and I understand how those tables relate.”_

That clarity will make database design much easier next week.
